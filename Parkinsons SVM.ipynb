```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn import manifold
from sklearn import preprocessing
```

```python
X = pd.read_csv("parkinsons.csv", header = 0,sep=',')
X.head()
```

```python
X.drop(labels = ["name"], axis = 1, inplace = True)
```

```python
X.head()
```

```python
X.columns
```

```python
y = X.loc[:, "status"]
X.drop(labels = ["status"], axis = 1, inplace =True)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 7)
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
import numpy as np

best_score = 0
best_C = 0.05
best_gamma = 0.001

for i in np.arange(0.05,2,0.05):
    for j in np.arange(0.001,0.1,0.001):
        model = SVC(C = i, gamma = j)
        model.fit(X_train, y_train)
        new_score = model.score(X_test, y_test)
        if(best_score < new_score):
            best_score = new_score
            best_C = i
            best_gamma = j
            
print("Best C, gamma and score is:", best_C, best_gamma, best_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 7)
print(type(X_train))
print(type(y_train))
```

```python
T_StandardScaler = preprocessing.StandardScaler().fit(X_train, y_train)
T_Normalizer = preprocessing.Normalizer().fit(X_train, y_train)
T_MaxAbsScaler = preprocessing.MaxAbsScaler().fit(X_train, y_train)
T_MinMaxScaler = preprocessing.MinMaxScaler().fit(X_train, y_train)
T_KernelCenterer = preprocessing.KernelCenterer().fit(X_train, y_train)

T_test_StandardScaler = T_StandardScaler.transform(X_test)
T_test_Normalizer = T_Normalizer.transform(X_test)
T_test_MaxAbsScaler = T_MaxAbsScaler.transform(X_test)
T_test_MinMaxScaler = T_MinMaxScaler.transform(X_test)
T_test_KernelCenterer = T_KernelCenterer.transform(X_test)

T_train_StandardScaler = T_StandardScaler.transform(X_train)
T_train_Normalizer = T_Normalizer.transform(X_train)
T_train_MaxAbsScaler = T_MaxAbsScaler.transform(X_train)
T_train_MinMaxScaler = T_MinMaxScaler.transform(X_train)
T_train_KernelCenterer = T_KernelCenterer.transform(X_train)

best_score = [0]
best_C = [0.05]
best_gamma = [0.001]

for i in np.arange(0.05,2.05,0.05):
    for j in np.arange(0.001,0.101,0.001):
        model = SVC(C = i, gamma = j)
        model.fit(T_train_StandardScaler, y_train)
        new_score = model.score(T_test_StandardScaler, y_test)
        if(best_score[0] < new_score):
            best_score[0] = new_score
            best_C[0] = i
            best_gamma[0] = j
            
print("Best C, gamma and score with preprocessing.StandardScaler is:", best_C[0], best_gamma[0], best_score[0])

best_score.append(0)
best_C.append(0.05)
best_gamma.append(0.001)

for i in np.arange(0.05,2.05,0.05):
    for j in np.arange(0.001,0.101,0.001):
        model = SVC(C = i, gamma = j)
        model.fit(T_train_Normalizer, y_train)
        new_score = model.score(T_test_Normalizer, y_test)
        if(best_score[1] < new_score):
            best_score[1] = new_score
            best_C[1] = i
            best_gamma[1] = j
            
print("Best C, gamma and score with preprocessing.Normalizer is:", best_C[1], best_gamma[1], best_score[1])

best_score.append(0)
best_C.append(0.05)
best_gamma.append(0.001)


for i in np.arange(0.05,2.05,0.05):
    for j in np.arange(0.001,0.101,0.001):
        model = SVC(C = i, gamma = j)
        model.fit(T_train_MaxAbsScaler, y_train)
        new_score = model.score(T_test_MaxAbsScaler, y_test)
        if(best_score[2] < new_score):
            best_score[2] = new_score
            best_C[2] = i
            best_gamma[2] = j
            
print("Best C, gamma and score with preprocessing.MaxAbsScaler is:", best_C[2], best_gamma[2], best_score[2])

best_score.append(0)
best_C.append(0.05)
best_gamma.append(0.001)

for i in np.arange(0.05,2,0.05):
    for j in np.arange(0.001,0.1,0.001):
        model = SVC(C = i, gamma = j)
        model.fit(T_train_MinMaxScaler, y_train)
        new_score = model.score(T_test_MinMaxScaler, y_test)
        if(best_score[3] < new_score):
            best_score[3] = new_score
            best_C[3] = i
            best_gamma[3] = j
            
print("Best C, gamma and score preprocessing.MinMaxScaler is:", best_C[3], best_gamma[3], best_score[3])

best_score.append(0)
best_C.append(0.05)
best_gamma.append(0.001)


for i in np.arange(0.05,2.05,0.05):
    for j in np.arange(0.001,0.101,0.001):
        model = SVC(C = i, gamma = j)
        model.fit(T_train_KernelCenterer, y_train)
        new_score = model.score(T_test_KernelCenterer, y_test)
        if(best_score[4] < new_score):
            best_score[4] = new_score
            best_C[4] = i
            best_gamma[4] = j
            
print("Best C, gamma and score preprocessing.KernelCenterer is:", best_C[4], best_gamma[4], best_score[4])
```

```python
best_preprocessor = best_score.index(max(best_score))
print("Best C, gamma and score After Preprocessing is:", best_C[best_preprocessor], best_gamma[best_preprocessor], best_score[best_preprocessor])
```
